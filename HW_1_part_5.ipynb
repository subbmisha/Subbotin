```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import pylab
from scipy.optimize import curve_fit
import scipy as sp
from matplotlib import pyplot as plt
```

```python
xdata = np.linspace(-3, 3, 500)
```

```python
y = lambda x: x**3 + 2*x**2 - 3*x + 2
```

```python
z = y(xdata) + np.random.randn(len(xdata))
print(type(z))
```

```python
pylab.plot(xdata, z)
```

```python
def func(x, a, b, c, d):
    return a * x**3 + b*x**2 + c*x + d
```

```python
popt, pcov = curve_fit(func, xdata, z)
```

```python
popt
```

```python
[a,b,c,d] = popt
```

```python
fnew = lambda x: a*x**3 + b*x**2 + c*x + d
```

```python
antifnew = lambda x: (-a*x**3 - b*x**2 - c*x - d)
```

```python
plt.plot(xdata, fnew(xdata))
```

```python
minim = sp.optimize.brent(fnew)
maxim = sp.optimize.brent(antifnew)
```

```python
plt.plot(xdata, z, 'g')
plt.plot(xdata, fnew(xdata), 'r')
plt.annotate('local min', xy=(minim, fnew(minim)), xytext=(minim, 20),arrowprops=dict(facecolor='black'))
plt.annotate('local max', xy=(maxim, fnew(maxim)), xytext=(maxim, 30),arrowprops=dict(facecolor='black'))
```
