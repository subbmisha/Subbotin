## Point 3.3

```python
%%time
np.linspace(13.0, 25.0, 15000000)
```

```python
def linspac(beg, end, nos):
    a = [beg + i*(end-beg)/(nos-1) for i in range(0, nos)]
    return a
```

```python
%%time
linspac(13.0, 25.0, 15000000)
```
