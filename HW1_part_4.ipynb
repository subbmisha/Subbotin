## Points 4.2-4.3

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pylab
from pylab import *
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 12.56, 1000)
y = np.sin(x)
pylab.plot(x, y, '-y', label='sin')

x = np.linspace(0, 12.56, 1000)
y = np.cos(x)
pylab.plot(x, y, '-g', label='cosin')

x = np.linspace(0, 12.56, 1000)
y = np.log(x)
pylab.plot(x, y, '-r', label='log')

x = np.linspace(-5, 1, 1000)
y = np.exp(x)
pylab.plot(x, y, '-b', label='exp')

pylab.legend(loc='upper right')
plt.xlabel('It is the X axis')
plt.ylabel('It is the Y axis')
plt.title('Functions $\sin(x)$ , $\cosin(x)$, $\ ln(x)$ , $\exp(x) $ ')


```
